After Pre_Processing:

```python
from sklearn.preprocessing import LabelEncoder

var = ['Group', 'GENDER', 'Age', 'EDUC', 'SES', 'MMSE', 'CDR', 'eTIV', 'nWBV',
       'ASF']

le = LabelEncoder()

for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
